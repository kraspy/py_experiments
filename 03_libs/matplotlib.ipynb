## Установка и импорт

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Настройка графиков

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)', color='purple', linestyle='--', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Настройка стиля графика')
plt.legend(loc='upper right')  # Размещение легенды
plt.grid(True)
plt.show()
````

## Полезные настройки и команды
```python
plt.figure(figsize=(12, 6))  # Размер графика
plt.style.use('seaborn')     # Установка стиля
plt.xticks(rotation=45)      # Поворот подписей оси X
plt.tight_layout()           # Автоматическое размещение графика
plt.clf()                    # Очистить текущий график
```

## Сохранение графика в файл
```python
plt.plot(x, y)
plt.title('Сохранение в файл')
plt.savefig('plot.png', dpi=300, bbox_inches='tight')  # Сохранить график
plt.show()
```

## Аннотации на графике

````python
plt.plot(x, y)
plt.annotate('Максимум', xy=(np.pi/2, 1), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Аннотация')
plt.grid(True)
plt.show()
````

## Основной график (line plot)

````python
x = np.linspace(0, 10, 100)  # 100 точек от 0 до 10
y = np.sin(x)

plt.plot(x, y, label='sin(x)', color='r', linewidth=2)  # Основной график
plt.xlabel('X-axis')    # Подпись оси X
plt.ylabel('Y-axis')    # Подпись оси Y
plt.title('Line Plot')  # Заголовок графика
plt.legend()            # Легенда
plt.grid(True)          # Сетка
plt.show()              # Отобразить график
````

## Диаграмма рассеяния (scatter plot)

````python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()  # Цветовая шкала
plt.title('Scatter Plot')
plt.show()
````

## Гистограмма (histogram)

````python
data = np.random.randn(1000)  # 1000 случайных значений

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Histogram')
plt.show()
````

## Столбчатая диаграмма (bar plot)

````python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values, color='lightcoral')
plt.title('Bar Plot')
plt.xlabel('Категории')
plt.ylabel('Значение')
plt.show()
````

## Круговая диаграмма (pie chart)

````python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [45, 30, 15, 10]
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
````

## График с несколькими подграфиками (subplots)

````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))

# Первый подграфик
plt.subplot(1, 2, 1)  # (rows, cols, index)
plt.plot(x, y1, color='r')
plt.title('sin(x)')

# Второй подграфик
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='b')
plt.title('cos(x)')

plt.tight_layout()
plt.show()
````

## 3D-график

````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x + y2))

ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Surface Plot')
plt.show()
````

## Boxplot

````python
data = [np.random.randn(100), np.random.randn(100) + 2, np.random.randn(100) - 2]
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Boxplot')
plt.xticks([1, 2, 3], ['Группа 1', 'Группа 2', 'Группа 3'])
plt.grid(True)
plt.show()
````

## Heatmap

````python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.show()
````

## Комбинирование графиков

````python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Комбинирование графиков')
plt.legend()
plt.grid(True)
plt.show()
````
